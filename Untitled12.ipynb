```python
thai_months = {
    'มกราคม': '01', 'กุมภาพันธ์': '02', 'มีนาคม': '03', 'เมษายน': '04',
    'พฤษภาคม': '05', 'มิถุนายน': '06', 'กรกฎาคม': '07', 'สิงหาคม': '08',
    'กันยายน': '09', 'ตุลาคม': '10', 'พฤศจิกายน': '11', 'ธันวาคม': '12'
}

# Convert month names to numbers
clean2019['month_num'] = clean2019['month'].map(thai_months)

# ฟังก์ชันสำหรับสร้างวันที่อย่างปลอดภัย
def create_date(row):
    try:
        return pd.to_datetime(f"{row['year']}-{row['month_num']}-{row['วันที่']}", format='%Y-%m-%d')
    except ValueError:
        return pd.NaT

# ใช้ฟังก์ชันในการสร้างคอลัมน์วันที่ใหม่
clean2019['date'] =clean2019.apply(create_date, axis=1)
```

```python
clean2019.dropna(subset=['date'],inplace=True)
```

```python
clean2019.reset_index(drop=True,inplace=True)
#drop=True เพื่อไม่สร้างคอลัมน์ใหม่สำหรับ index เดิม.
```

```python
# ฟังก์ชันในการแยกตัวเลขในวงเล็บและแทนค่าลงไปที่ตำแหน่งเดิม
def extract_and_replace(value):
    if isinstance(value, str):  # ตรวจสอบว่าค่าเป็นสตริงหรือไม่
        match = re.search(r'\(([^)]+)\)', value)
        if match:
            return match.group(1)
    return value  # Return ค่าดั้งเดิมถ้าไม่ใช่สตริงหรือไม่มีตัวเลขในวงเล็บ
```

```python
max2020 = max2020.applymap(extract_and_replace)
max2019 = max2019.applymap(extract_and_replace)
max2021 = max2021.applymap(extract_and_replace)
max2022 = max2022.applymap(extract_and_replace)
max2023 = max2023.applymap(extract_and_replace)
```
