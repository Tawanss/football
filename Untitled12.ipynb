```python
df = pd.read_excel(file_path,header=1,)
```

```python
df = df.rename(columns={'Unnamed: 0': 'วันที่'})
```

```python
x = df.transpose()
```

```python
x.reset_index(inplace=True)
```

```python
x.drop(x.index[0],inplace=True)
```

```python
x
```

```python
x.columns = range(x.shape[1])
```

```python
x.rename(columns={0: 'month'},inplace=True)
```

```python
x.drop(columns=[32,33],inplace=True)
```

```python
clean2019 = x.set_index(['month']).stack().reset_index()
```

```python
clean2019.rename(columns={'level_1':'วันที่', 0 : 'น้ำระเหย'},inplace = True)
```

```python
clean2019['year'] = 2019
```

```python
clean2019
```

```python
thai_months = {
    'มกราคม': '01', 'กุมภาพันธ์': '02', 'มีนาคม': '03', 'เมษายน': '04',
    'พฤษภาคม': '05', 'มิถุนายน': '06', 'กรกฎาคม': '07', 'สิงหาคม': '08',
    'กันยายน': '09', 'ตุลาคม': '10', 'พฤศจิกายน': '11', 'ธันวาคม': '12'
}

# Convert month names to numbers
clean2019['month_num'] = clean2019['month'].map(thai_months)

# ฟังก์ชันสำหรับสร้างวันที่อย่างปลอดภัย
def create_date(row):
    try:
        return pd.to_datetime(f"{row['year']}-{row['month_num']}-{row['วันที่']}", format='%Y-%m-%d')
    except ValueError:
        return pd.NaT

# ใช้ฟังก์ชันในการสร้างคอลัมน์วันที่ใหม่
clean2019['date'] =clean2019.apply(create_date, axis=1)
```

```python
clean2019.dropna(subset=['date'],inplace=True)
```

```python
clean2019.reset_index(drop=True,inplace=True)
#drop=True เพื่อไม่สร้างคอลัมน์ใหม่สำหรับ index เดิม.
```
