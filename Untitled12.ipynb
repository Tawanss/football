# time series

```python
from prophet import Prophet
import matplotlib.pyplot as plt
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
time_series = merge[['date','อุณหภูมิสูงสุด','อุณหภูมิต่ำสุด']]
```

```python
time_series.replace('-', pd.NA, inplace=True)
```

```python
time_series.isnull().any() #ไม่มีnull
```

```python
time_series
```

```python
#อุณหภูมิสูงสุด
df = time_series[['date','อุณหภูมิสูงสุด']]
df.columns=['ds','y']
```

```python
df['y'] = df['y'].astype(float)
```

```python
df.set_index('ds').plot(figsize=(16,6));
```

```python
model=Prophet()
```

```python
model.fit(df)
```

```python
#predict
future = model.make_future_dataframe(periods=100, freq='D')
future
```

```python
forecast = model.predict(future)
```

```python
forecast
```

```python
dffor = forecast[['ds','yhat','yhat_lower','yhat_upper']]
```

```python
summary = pd.merge(df,dffor,left_on='ds',right_on='ds')
```

```python
summary
```

```python
fig1 = model.plot(forecast)
plt.show()
```

```python
# แสดงองค์ประกอบของการพยากรณ์
fig2 = model.plot_components(forecast)
plt.show()
```

```python
y_true = summary['y']
y_pred = summary['yhat']
```

```python
#ประสิทธิภาพโมเดล
mae = mean_absolute_error(y_true, y_pred)
mse = mean_squared_error(y_true, y_pred)
rmse = np.sqrt(mse)
mape = np.mean(np.abs((y_true - y_pred) / y_true)) * 100

print(f'MAE: {mae}')
print(f'MSE: {mse}')
print(f'RMSE: {rmse}')
print(f'MAPE: {mape}%')
```
